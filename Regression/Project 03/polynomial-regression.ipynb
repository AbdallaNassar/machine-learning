```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df =pd.read_csv("polynomial-regression.csv")
df.head()
```

```python
x=df.iloc[:,: -1]
y=df.iloc[:,-1]
model=LinearRegression()
model.fit(x,y)
print(model.score(x,y))
#very bad :(
plt.scatter(x,y)
plt.plot(x,model.predict(x) ,color="red")
plt.show()
```

```python
poly=PolynomialFeatures(degree=2)
p_x=poly.fit_transform(x)
model=LinearRegression()
model.fit(p_x,y)
print(model.score(p_x,y))
plt.scatter(x,y)
plt.plot(x,model.predict(p_x) ,color="red")
plt.show()
```

```python
poly=PolynomialFeatures(degree=3)
p_x=poly.fit_transform(x)
model=LinearRegression()
model.fit(p_x,y)
print(model.score(p_x,y))
plt.scatter(x,y)
plt.plot(x,model.predict(p_x) ,color="red")
plt.show()
```

```python
poly=PolynomialFeatures(degree=4)
p_x=poly.fit_transform(x)
model=LinearRegression()
model.fit(p_x,y)
print(model.score(p_x,y))
plt.scatter(x,y)
plt.plot(x,model.predict(p_x) ,color="red")
plt.show()# it is best -__-
```
